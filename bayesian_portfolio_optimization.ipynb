```python
import numpy as np
import yfinance as yf
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
tickers = ["GOOGL", "NVDA", "META", "PLTR", "WFC",
           "DIS", "LUV", "PFE", "COKE", "CAT"]

start = "2021-01-01"
end = "2026-01-01"
T = 252
```

```python
data = yf.download(tickers, start=start, end=end)["Close"]
```

```python
def get_means(data, T):
  returns = data.pct_change().dropna()*100
  mu_0 = returns.mean()
  mu_hat = returns.iloc[-T:].mean()

  return mu_0, mu_hat

def tau(mu_0, mu_hat):
  return np.sqrt(np.var(mu_hat)/np.var(mu_0))
```

```python
mu_0, mu_hat = get_means(data, T=T)
tau = tau(mu_0, mu_hat)
```
