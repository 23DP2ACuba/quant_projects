```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
S = 101.15
K = 98.01
vol = 0.0991
r = 0.01
N = 1
M = 100000
m_v = 3.76
T = ((datetime.date(2026, 2, 22) - datetime.date.today()).days+1) / 365
print(T)
```

```python
dT = T/N
nudT = (r - 0.5*vol**2)*dT
sdT = vol*np.sqrt(dT)
lnS = np.log(S)
W = np.random.normal(size=(N, M))
delta_lnS = nudT + sdT * W
lnS_t = lnS + np.cumsum(delta_lnS, axis=0)
lnS_t = np.concatenate((np.full(shape=(1, M), fill_value=lnS), lnS_t))
```

```python
ST = np.exp(lnS_t)
CT = np.maximum(ST-K, 0)
C0 = np.exp(-r*T)*np.sum(CT[-1])/M
sigma= np.sqrt(np.sum((CT[-1] - C0)**2) / (M-1))
SE = sigma/np.sqrt(M)
```

```python
print(f"Call: {C0}, with SE: {SE}")
```

```python
import scipy as si
```

```python
x1 = np.linspace(C0-3*SE, C0+SE, 100)
x2 = np.linspace(C0-2*SE, C0+2*SE, 100)
x3 = np.linspace(C0-SE, C0+3*SE, 100)

s1 = si.stats.norm.pdf(x1, C0, SE)
s2 = si.stats.norm.pdf(x2, C0, SE)
s3 = si.stats.norm.pdf(x3, C0, SE)

plt.fill_between(x1, s1, color="red")
plt.fill_between(x2, s2, color="blue")
plt.fill_between(x3, s3, color="green")

plt.plot([C0, C0], [0, max(s2)*1.1], "k")
plt.plot([m_v, m_v], [0, max(s2)**1.1])
```
