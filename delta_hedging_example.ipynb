```python
import numpy as np
from scipy import stats
```

```python
S_0 = 100
K = 100
T = 30/365
r = 0.01
sigma = 0.2

S_1 = 101
```

```python
def get_position(K, S_0, S_1, sigma, r, T):
  d1 = (np.log(S_0/K) + (r+sigma**2/2)*T) / (sigma*np.sqrt(T))
  delta = stats.norm.cdf(d1)
  print(f"Delta: {delta}")
  del_S = S_1-S_0
  pos = -1 * delta * del_S
  return (pos, d1)
```

```python
pos, d1 = get_position(K, S_0, S_1, sigma, r, T)
print(pos)
```

```python
def get_gamma(d1, S_0, S_1, sigma, T):
  Nd1 = 1/np.sqrt(2*np.pi)* np.exp(-d1**2/2)
  g = Nd1 / (S_0 * sigma * np.sqrt(T))
  res_pl = (S_1 - S_0) * g * 0.5
  print(f"P/L including gamma: {res_pl}")
  
  return g
  
```

```python
get_gamma(d1, S_0, S_1, sigma, T)
```
